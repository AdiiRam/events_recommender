```python
import matplotlib
import pandas as pd
%matplotlib inline
```

```python
# Read the CSV File into a Dataframe 
events_dataset = './events_recommender/resources/event_dataset.csv'
```

```python
events_df = pd.read_csv(events_dataset)
events_df.info()
```

### Total No of Events 

```python
events_df['event_edition_id'].count()
```

### Total No of Non Null and Non Empty Records per Column Present 

```python
events_df.count()
```

#### Grouping the Events by Category 

```python
events_df.groupby('event_category')['event_edition_id'].count().plot.pie(autopct='%1.1f%%',startangle=180)
```

### Grouping the Events by Registration Type 

```python
events_df.groupby('registration_type')['event_edition_id'].count().plot.pie(autopct='%1.1f%%',startangle=90)
```

### Grouping the Events by city

```python
events_df.groupby('city')['event_edition_id'].count().plot.bar(width=1,log=True)
```
